```python
import random
import pandas as pd
import numpy

import faker 
from datetime import datetime, timedelta


fake = faker.Faker()

def generate_transaction(num_records=10000):

    data = []
    start_date = datetime.now() - timedelta(days=730)

    for i in range(1, num_records+1):

        #category = random.choice(list(expense_categories.keys())) 
        #Trying weightage category
        weighted_categories = random.choices(
            list(expense_categories.keys()), 
            weights=[20,15,10,10,10,7,7,5,5,5,3,3],  
            k=1
        )[0]
        
        note = random.choice(expense_categories[weighted_categories])
        if note in item_amounts:
            low, high = item_amounts[note]
        else:       
            low, high = category_amounts[weighted_categories]
        amount = round(random.uniform(low, high), 2)
        days_offset = random.randint(0, 730)
        date = start_date + timedelta(days=days_offset)

        data.append([
            i, 
            date.strftime("%Y-%m-%d"),
            amount,
            note,
            weighted_categories
        ])

    df = pd.DataFrame(data, columns=["transaction_id", "date", "amount", "note", "category"])
    return df

df = generate_transaction(10000)
df.to_csv("transactions.csv", index=False)
print("Done")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.preprocessing import FunctionTransformer
import numpy as numpy
```

```python
df = gen_df.copy()
```

```python
y = df['category']
X = df[['amount', 'date', 'note']]
```

```python
def extract_date_features(d):
    s = pd.to_datetime(d.iloc[:, 0])
    return pd.DataFrame({
        'month': s.dt.month.astype('int8'),
        'dow': s.dt.dayofweek.astype('int8'),
        'is_weekend': (s.dt.dayofweek >= 5).astype('int8')
    })
```

```python
date_feat = Pipeline(steps=[
    ('extract', FunctionTransformer(extract_date_features, validate=False)),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocess = ColumnTransformer(transformers=[
    ('num', StandardScaler(), ['amount']),
    ('date', date_feat, ['date']),
    ('text', TfidfVectorizer(max_features=5000, ngram_range=(1,2)), 'note')
])
```

```python
clf = Pipeline(steps=[
    ('prep', preprocess),
    ('model', LogisticRegression(max_iter=200, multi_class='multinomial', n_jobs=-1))
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Macro F1:', f1_score(y_test, y_pred, average='macro'))
print(classification_report(y_test, y_pred))
```

```python
examples = pd.DataFrame([
    {"note": "Chai tapri",            "amount": 20,    "date": pd.Timestamp.today()},
    {"note": "Street food",           "amount": 120,   "date": pd.Timestamp.today()},
    {"note": "Uber ride",             "amount": 300,   "date": pd.Timestamp.today()},
    {"note": "Paid rent",             "amount": 18000, "date": pd.Timestamp.today()},
    {"note": "Netflix subscription",  "amount": 499,   "date": pd.Timestamp.today()},
    {"note": "Udemy course",          "amount": 799,   "date": pd.Timestamp.today()},
    {"note": "Flight booking",        "amount": 12000, "date": pd.Timestamp.today()},
])
```

```python
pred = clf.predict(examples)
proba = clf.predict_proba(examples)
labels = clf.named_steps["model"].classes_

topk = 3
for i, row in examples.iterrows():
    idx = np.argsort(proba[i])[::-1][:topk]
    print(f"{row['note']} | {row['amount']} -> {pred[i]}")
    for j in idx:
        print(f"  {labels[j]}: {proba[i][j]:.2f}")
    print()

```
